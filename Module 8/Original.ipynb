```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('/content/drive/MyDrive/Part 2/Part 2/Part 2training_images/real_00004.jpg')

Face_Classifier = cv2.CascadeClassifier('/content/drive/MyDrive/Part 2/Part 2/haar-cascade-files-master/Frontal Face/Frontalface_alt_tree.xml')
IMAGE = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


Face = Face_Classifier.detectMultiScale(IMAGE, 1.0485258, 6)


if Face is ():
    print("No faces found")

    
for (x,y,w,h) in Face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,250,250), 10)
    plt.imshow(img)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread('/content/drive/MyDrive/Part 2/Part 2/Part 2training_images/real_00006.jpg')

Face_Classifier = cv2.CascadeClassifier('/content/drive/MyDrive/Part 2/Part 2/haar-cascade-files-master/Frontal Face/Frontalface_alt_tree.xml')
IMAGE = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

''' Our classifier returns the ROI of the detected face as a tuple, 
It stores the top left coordinate and the bottom right coordiantes'''

Face = Face_Classifier.detectMultiScale(IMAGE, 1.0485258, 6)
'''When no faces detected, face_classifier returns and empty tuple'''

if Face is ():
    print("No faces found")
'''We iterate through our faces array and draw a rectangle over each face in faces'''
for (x,y,w,h) in Face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (250,0,222), 5)
    print('x:',x,'y:',y,'w:',w,'h:',h)
    plt.imshow(img)
```
